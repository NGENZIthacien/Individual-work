**Exercice 3 : Grouping & Aggregating Practice**

 14.	Calculate the total QuantitySold for each Category.

```python
df.groupby('Category')['QuantitySold'].sum()
```

15.	Calculate the average TotalRevenue for each Region, sorted from highest to lowest.

```python
df['TotalRevenue'] = df['UnitPrice'] * df['QuantitySold']
df.groupby('Region')['TotalRevenue'].mean().sort_values(ascending=False)
```

16.	Using .agg() with named aggregation, calculate, for each Region: total revenue, number of sales, and the maximum single-sale revenue.

```python
df.groupby('Region').agg(
    TotalRevenue=('TotalRevenue', 'sum'),
    NumberOfSales=('TotalRevenue', 'count'),
    MaxSingleSale=('TotalRevenue', 'max')
)
```

17.	Find out which Category appears most frequently in the dataset using value_counts().

18.	Group by both Category and Region, and find the total revenue for each combination. Use .reset_index() so the result is a flat table.

```python
df.groupby(['Category', 'Region'])['TotalRevenue'].sum().reset_index()
```

**Exercise 4 : The Manager's Questions**

19.	“What is our total revenue across the entire dataset?” — Calculate a single overall total.

```python
total_revenue = df['TotalRevenue'].sum()

print(total_revenue)
```

20.	“Which single product generated the most total revenue, and how much was it?” - Identify the top product and its exact revenue figure.

```python
revenue_by_product = df.groupby('Product')['TotalRevenue'].sum()

top_product = revenue_by_product.sort_values(ascending=False)

print(top_product.head(1))
```

21.	“Are sales fairly even across all four regions, or is one region falling behind?” - Calculate total revenue per region and comment on what you notice.

```python
revenue_by_region = df.groupby('Region')['TotalRevenue'].sum().sort_values(ascending=False)

print(revenue_by_region)
```

22.	“What were our 3 best individual sales days, by total revenue?” — Group by Date and find the top 3 days.

```python
revenue_by_day = df.groupby('Date')['TotalRevenue'].sum()

print(revenue_by_day.sort_values(ascending=False).head(3))
```

23.	“Out of all our sales, what percentage came from the Dairy category?” - Hint: you will need total Dairy revenue divided by total overall revenue, multiplied by 100.

```python
dairy_revenue = df[df['Category'] == 'Dairy']['TotalRevenue'].sum()

total_revenue = df['TotalRevenue'].sum()

dairy_percentage = (dairy_revenue / total_revenue) * 100

print(f'Dairy made up {dairy_percentage:.1f}% of total revenue')
```
